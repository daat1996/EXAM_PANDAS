(9) DataFrame의 원소들 모두 출력 단, 2가지 방법 사용

```python
mdf=read_csv(dataFile)     # 복구
mdf
```

```python
for a in range(len(mdf.index)):      #len 대신에 mdf.shape[0] 가능
    print()
    for b in range(len(mdf.columns)):      # len 대신에 mdf.shape[1] 가능
        print(mdf.iloc[a][b],end=' ')
```

```python
for row in mdf.index:
    print()
    for col in mdf.columns:
        print(mdf.loc[row][col], end=' ')
```
