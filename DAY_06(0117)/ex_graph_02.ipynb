(2) 데이터 준비 <hr>

```python
# 2D 그래프를 그리기 위한 데이터 준비
x = range(0,100,10)              # 0 ~ 99
y = [value **2 for value in x]
x,y
```

### x,y축 레이블 설정 및 눈금 설정
- x, y 축 데이터에 대한 설명 => plt.xlabel(), plt.ylabel()
- 눈금 간격, 출력 형태 설정 => plt.xticks(), plt.yticks(), plt.tick_param()

```python
# 그래프 그리기
plotList=plt.plot(x,y,'g^--',x2,y2, 'ro-')
plt.title('[테스트]')

# x축, y축 레이블
plt.xlabel('값',loc='right',fontsize='x-large',color='red')
plt.ylabel('결과',loc='top',rotation='horizontal')

# x, y 축 눈금 설정
plt.xticks(ticks=[15,30,45,60,75], labels=['A','B','C','D','E'], color='red')
plt.yticks(color='blue')

# 범례 설정
plt.legend(['green 결과','red 결과'], loc='center')

# 그래프 안에 텍스트 출력
plt.text(2,2000,'Happy New Year')
```

### 범례 설정
- 그래프 내에 위치 지정 후 텍스트 추가 설정 => plt.text(시작x, 시작y, 출력텍스트)

```python
# 그래프 그리기
plotList=plt.plot(x,y,'g^--',x2,y2, 'ro-')
plt.title('[테스트]')

# x축, y축 레이블
plt.xlabel('값',loc='right',fontsize='x-large',color='red')
plt.ylabel('결과',loc='top',rotation='horizontal')

# x, y 축 눈금 설정
plt.xticks(ticks=[15,30,45,60,75], labels=['A','B','C','D','E'], color='red')
plt.yticks(color='blue')

# 범례 설정
plt.legend(['green 결과','red 결과'], loc='center')

# 그래프 안에 텍스트 출력
plt.text(20,2000,'Happy New Year', backgroundcolor='yellow',color='red',va='top')

box={'facecolor':'.75','edgecolor':'k','boxstyle':'round','alpha':0.5}
plt.text(40,6000,'Happy New Year', bbox=box)
```

### 화살표 설정
- 그래프에 특정 지점에 화살표 표기
- text와 함께 종종 사용됨
- plt.annotate()

```python
# 그래프 그리기
plt.figure(figsize=(10,5))        # 사이즈 변경 figsize:(단위 인치)
plotList=plt.plot(x,y,'g^--',x2,y2, 'ro-')
plt.title('[테스트]')

# x축, y축 레이블
plt.xlabel('값',loc='right',fontsize='x-large',color='red')
plt.ylabel('결과',loc='top',rotation='horizontal')

# x, y 축 눈금 설정
plt.xticks(ticks=[15,30,45,60,75], labels=['A','B','C','D','E'], color='red')
plt.yticks(color='blue')

# 범례 설정
plt.legend(['green 결과','red 결과'], loc='center')

# 화살표 설정
plt.annotate('Brackmard minimum', 
            ha = 'center', va = 'bottom',
            xytext = (20,3000),
            xy = (70,6000),
            arrowprops={'facecolor':'black','shrink':1})
```

```python
x= np.linspace(0, 6, 1024)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.xlabel('X')
plt.ylabel('Y', rotation='horizontal')

plt.plot(x, y1, c='k',lw=3, label ='sin(X)')
plt.plot(x, y2, c='.5',lw=3, ls='--', label ='cos(X)')
plt.show()
```
