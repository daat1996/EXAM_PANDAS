```python
def histogram_intersection(a, b):
    v = np.minimum(a, b).sum().round(decimals=1)
    return v
df = pd.DataFrame([(.2, .3), (.0, .6), (.6, .0), (.2, .1)],
                  columns=['dogs', 'cats'])
df.corr(method=histogram_intersection)
```

```python
df.corr()
```

```python
df = pd.DataFrame([(1, 1), (2, np.nan), (np.nan, 3), (4, 4)],
                  columns=['dogs', 'cats'])
df.corr(min_periods=3)
```

```python
def scatterplot(sr1, sr2):
    '''
    parameters:
    sr1, sr2 : Series
    '''
    plt.scatter(sr1, sr2)
    plt.title(f'{sr1.name} - {sr2.name} [{round(corrDF[sr1.name][sr2.name],2)}]')
    plt.xlabel(sr1.name)
    plt.ylabel(sr2.name)
    plt.show()
```

```python
fig, ax = plt.subplots()

ax1 = fig.add_subplot(311)
ax1.scatter(mpgDF['mpg'], mpgDF['weight'])
ax1.set_title('mpg-weight')

ax2 = fig.add_subplot(312)
ax2.scatter(mpgDF['mpg'], mpgDF['weight'])
ax2.set_title('mpg-weight')

ax3 = fig.add_subplot(313)
ax3.scatter(mpgDF['mpg'], mpgDF['cylinders'])
ax3.set_title('mpg-cylinders')

plt.tight_layout()
plt.show()
```

```python
# First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

# Create just a figure and only one subplot
fig, ax = plt.subplots()       # 1 행 1열
ax.plot(x, y)
ax.set_title('Simple plot')
```

```python
# Create two subplots and unpack the output array immediately
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)   # sharey : y축을 공유하겠다.
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)
```

```python
# Create four polar axes and access them through the returned array
fig, axs = plt.subplots(2, 2)    # 먼저 공간을 만들어 놓는다 
axs[0, 0].plot(x, y)
axs[1, 1].scatter(x, y)
```

```python
# Share a X axis with each column of subplots     x축 값 공유
plt.subplots(2, 2, sharex='col')
```

```python
# Share a Y axis with each row of subplots        y축 값 공유
plt.subplots(2, 2, sharey='row')
```

```python
# Share both X and Y axes with all subplots       둘다 공유
plt.subplots(2, 2, sharex='all', sharey='all')
```

```python
# Note that this is the same as         얘도 둘다 공유
plt.subplots(2, 2, sharex=True, sharey=True)
```

```python
# Create figure number 10 with a single subplot
# and clears it if it already exists.
fig, ax = plt.subplots(num=10, clear=True)
```
